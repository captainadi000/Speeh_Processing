```python
import numpy as np
from glob import glob
import librosa
import librosa.display
import IPython.display as ipd
from IPython.display import Audio, display
from itertools import cycle
import pygame
import soundfile as sf
```

Q1)  
Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.

```python
audio =  r"/content/adithya.wav"
ipd.Audio(audio)
y, sr = librosa.load(audio)
librosa.display.waveshow(y)
```

Q2) Observe the length & magnitude range of the signal from theplot. Observing the plot, relate to the spoken words and silence between the words.

```python
duration = len(y)/sr
print(f"Duration: {duration} seconds")
min_mag=min(y)
max_mag=max(y)
mag_range=max_mag-min_mag
print(f"Magnitude Range: {mag_range}")
```
